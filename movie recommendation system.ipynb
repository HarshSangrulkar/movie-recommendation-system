```python
movies=movies[['id', 'title', 'overview', 'genre']]
movies
```

```python
movies['tags'] = movies['overview']+movies['genre']
movies
```

```python
new_data  = movies.drop(columns=['overview', 'genre'])
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000, stop_words='english')
cv
```

```python
vector=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
similarity
```

```python
def recommend(movies):
    index=new_data[new_data['title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].title)
```

```python
#testing 1

recommend('Iron Man')
```

```python
#testing 2

recommend('Dilwale Dulhania Le Jayenge')
```

```python
#testing 3

recommend('Captain America')
```

```python
def recommend_movies(user_input, movies, similarity_matrix, data):
    index = movies[movies['title'] == user_input].index
    if not index.empty:
        index = index[0]
        distance = sorted(list(enumerate(similarity_matrix[index])), reverse=True, key=lambda vector: vector[1])
        recommended_movies = [data.iloc[i[0]].title for i in distance[1:6]]
        return recommended_movies
    else:
        return None

user_input = input("Enter a movie title: ")
recommended_movies = recommend_movies(user_input, new_data, similarity, new_data)

if recommended_movies:
    print(f"\nRecommended movies for '{user_input}':")
    for movie in recommended_movies:
        print(movie)
else:
    print(f"No movies found for '{user_input}'. Please check your input.")
```
